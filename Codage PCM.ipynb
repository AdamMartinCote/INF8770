Exemple de codage PCM simulé

```python
import numpy as np
import matplotlib.pyplot as plt
```

Génération d'un signal et affichage. Puisqu'on est déjà en numérique, on ré-échantillonera ce signal.

```python
x = np.arange(1.0,360.0, 0.5)/360*2*np.pi
signal = abs(np.sin(x) + np.sin(2*x) + np.cos(3*x))*100
plt.plot(x, signal)
plt.show()
```

Ré-échantillonage du signal avec 2 bits (4 niveaux)

```python
nbits = 2
tailleniveau = max(signal)/2**nbits

quantification = np.ceil(signal/tailleniveau)
signalPCM = quantification * tailleniveau 
    
plt.plot(x, signalPCM)
plt.show()
```

Ré-échantillonage du signal avec 4 bits (16 niveaux)

```python
nbits = 4
tailleniveau = max(signal)/2**nbits

quantification = np.ceil(signal/tailleniveau)
signalPCM = quantification * tailleniveau 
    
plt.plot(x, signalPCM)
plt.show()
```

Ré-échantillonage du signal avec 6 bits (64 niveaux)

```python
nbits = 6
tailleniveau = max(signal)/2**nbits

quantification = np.ceil(signal/tailleniveau)
signalPCM = quantification * tailleniveau 
    
plt.plot(x, signalPCM)
plt.show()
```

On peut aussi changer la fréquence d'échantillonage. Par exemple, à toutes les 8 valeurs

```python
nbits = 6
tauxech = 8
tailleniveau = max(signal)/2**nbits
signalPCM = [0] * len(signal)
for i in range(int(len(signal)/tauxech)):
    quantification = np.ceil(signal[i*tauxech]/tailleniveau)
    signalPCM[tauxech*i:tauxech*i+tauxech] = [quantification * tailleniveau]*tauxech 
    
plt.plot(x, signalPCM)
plt.show()
```

à toutes les 16 valeurs

```python
nbits = 6
tauxech = 16
tailleniveau = max(signal)/2**nbits
signalPCM = [0] * len(signal)
for i in range(int(len(signal)/tauxech)):
    quantification = np.ceil(signal[i*tauxech]/tailleniveau)
    signalPCM[tauxech*i:tauxech*i+tauxech] = [quantification * tailleniveau]*tauxech 
    
plt.plot(x, signalPCM)
plt.show()
```

à toutes les 64 valeurs. On perd la forme du signal. Pas assez d'échantillons... Théorème de Nyquist...

```python
nbits = 6
tauxech = 64
tailleniveau = max(signal)/2**nbits
signalPCM = [0] * len(signal)
for i in range(int(len(signal)/tauxech)):
    quantification = np.ceil(signal[i*tauxech]/tailleniveau)
    signalPCM[tauxech*i:tauxech*i+tauxech] = [quantification * tailleniveau]*tauxech 
    
plt.plot(x, signalPCM)
plt.show()
```
