Effet de la loi mu sur un signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

Génération d'un signal et affichage.

```python
x = np.arange(1.0,360.0, 0.5)/360*2*np.pi
signal = np.sin(x) + np.sin(2*x) + np.cos(3*x)
smax = max(signal)
smin = min(signal)
signal = (2*(signal - smin)/(smax-smin))-1
plt.plot(x, signal)
plt.show()
```

Distribution des valeurs du signal. On voit qu'il y a des valeurs grandes et petites.

```python
hist, intervalles = np.histogram(signal, bins=16)
plt.bar(intervalles[:-1], hist, width = 1)
plt.xlim(min(intervalles), max(intervalles))
plt.show() 
```

Application de la loi mu sur ce signal. Le signal est transformé pour réduire le nombre de valeurs petits qui sont plus sujettent à être perdues à cause du bruit. 

```python
signalloimu = np.sign(signal)*(np.log(1+(255*abs(signal)))/np.log(255+1))
plt.plot(x, signalloimu)
plt.show()
```

Distribution des valeurs du signal après la loi mu. Il y a moins de petites valeurs. Le signal peut être plus facilement transmis dans un canal bruité.

```python
hist, intervalles = np.histogram(signalloimu, bins=16)
plt.bar(intervalles[:-1], hist, width = 1)
plt.xlim(min(intervalles), max(intervalles))
plt.show() 
```

On récupère le signal d'origine par la transformée inverse.

```python
invsignalloimu=np.sign(signalloimu)*(1.0/255.0)*(((1.0+255.0)**abs(signalloimu))-1)
plt.plot(x, invsignalloimu)
plt.show()
```
