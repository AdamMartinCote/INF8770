Exemple d'incrustation

```python
import numpy as np
import matplotlib.pyplot as py
```

Lecture de l'image et affichage de celle-ci. 

```python
imagelue = py.imread('320px-Girl_in_front_of_a_green_background.jpg')
image=imagelue.astype('float')
py.imshow(imagelue)
py.show()
```

Application d'un algorithme d'incrustation

```python
k0 = 1.4
k1 = 1.0
k2 = 40.0
Rouge = image[:,:,0]
Vert = image[:,:,1]
Bleu = image[:,:,2]

imagesortie = np.copy(image)

#Détection du vert
alpha = k0*Bleu-k1*Vert+k2;

#Application du masque
Rouge[alpha<=0] = 0
Vert[alpha<=0] = 0
Bleu[alpha<=0] = 0

imagesortie[:,:,0] = Rouge
imagesortie[:,:,1] = Vert
imagesortie[:,:,2] = Bleu
```

```python
py.imshow(imagesortie.astype('uint8'))
py.show()
```
